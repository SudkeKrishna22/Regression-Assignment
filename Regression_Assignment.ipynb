**Question 1:** What is Simple Linear Regression?  
**Answer:**  
Simple Linear Regression is a basic model that shows how one variable affects another. It finds a straight-line relationship between X (input) and Y (output) using the formula **Y = mX + c**, where *m* is the slope and *c* is the intercept.


**Question 2:** What are the key assumptions of Simple Linear Regression?  
**Answer:**  
1. The relation between X and Y should be linear.  
2. The data points should be independent.  
3. The errors (residuals) should have equal variance.  
4. Errors should be normally distributed.  
5. There shouldn’t be any big outliers.

**Question 3:** What does the coefficient m represent in the equation Y = mX + c?  
**Answer:**  
The coefficient *m* is the slope — it tells how much Y changes when X increases by 1 unit.


**Question 4:** What does the intercept c represent in the equation Y = mX + c?  
**Answer:**  
The intercept *c* is the value of Y when X = 0. Basically, it’s where the line cuts the Y-axis.


**Question 5:** How do we calculate the slope m in Simple Linear Regression?  
**Answer:**  
We use this formula:  
\[
m = \frac{n(\sum XY) - (\sum X)(\sum Y)}{n(\sum X^2) - (\sum X)^2}
\]  
It basically gives the line that best fits the data.


**Question 6:** What is the purpose of the least squares method in Simple Linear Regression?  
**Answer:**  
The least squares method tries to make the total error (difference between real and predicted values) as small as possible by minimizing the squared differences.


**Question 7:** How is the coefficient of determination (R²) interpreted in Simple Linear Regression?  
**Answer:**  
R² tells how well the line fits the data.  
- If R² = 1 → perfect fit  
- If R² = 0 → no relation at all  
It shows how much of the change in Y is explained by X.


**Question 8:** What is Multiple Linear Regression?  
**Answer:**  
Multiple Linear Regression is like simple regression but with more than one X. The equation looks like:  
*Y = b₀ + b₁X₁ + b₂X₂ + ... + bₙXₙ* , It shows how several variables together affect Y.


**Question 9:** What is the main difference between Simple and Multiple Linear Regression?  
**Answer:**  
Simple Linear Regression has only one input variable, while Multiple Linear Regression has two or more inputs affecting the output.


**Question 10:** What are the key assumptions of Multiple Linear Regression?  
**Answer:**  
1. Relationship between variables is linear.  
2. Errors are independent.  
3. Errors have equal variance (no heteroscedasticity).  
4. Variables shouldn’t be too correlated (no multicollinearity).  
5. Residuals are normally distributed.


**Question 11:** What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?  
**Answer:**  
Heteroscedasticity means the errors don’t have equal variance — they spread unevenly. It makes the model less reliable and can mess up standard errors and confidence intervals.

**Question 12:** How can you improve a Multiple Linear Regression model with high multicollinearity?  
**Answer:**  
You can:  
- Drop one of the correlated variables.  
- Use PCA (Principal Component Analysis).  
- Use Ridge or Lasso regression to handle correlation better.

**Question 13:** What are some common techniques for transforming categorical variables for use in regression models?  
**Answer:**  
We can convert categories into numbers using:  
- One-Hot Encoding 
- Label Encoding 
- Ordinal Encoding  
This helps the model understand categorical data.

**Question 14:** What is the role of interaction terms in Multiple Linear Regression?  
**Answer:**  
Interaction terms show how two variables together affect the output. It’s like saying “the effect of X₁ depends on X₂.”

**Question 15:** How can the interpretation of intercept differ between Simple and Multiple Linear Regression?  
**Answer:**  
- In simple regression: intercept = Y when X = 0.  
- In multiple regression: intercept = Y when all Xs are 0, which doesn’t always make sense in real life.

**Question 16:** What is the significance of the slope in regression analysis, and how does it affect predictions?  
**Answer:**  
The slope shows how much the dependent variable changes when the independent variable increases by 1.  
Positive slope → Y increases,  
Negative slope → Y decreases.

**Question 17:** How does the intercept in a regression model provide context for the relationship between variables?  
**Answer:**  
The intercept acts like a starting point. It gives the predicted Y value when all inputs are zero, helping us understand the base level.

**Question 18:** What are the limitations of using R² as a sole measure of model performance?  
**Answer:**  
- R² always increases with more features, even useless ones.  
- It doesn’t show overfitting.  
- It can’t tell if predictions are good on new data.

**Question 19:** How would you interpret a large standard error for a regression coefficient?  
**Answer:**  
A big standard error means the coefficient isn’t stable — the model isn’t sure how much that variable really affects Y.

**Question 20:** How can heteroscedasticity be identified in residual plots, and why is it important to address it?  
**Answer:**  
If residuals form a pattern (like a funnel shape), it means there’s heteroscedasticity. It’s bad because it makes the model’s predictions and confidence intervals unreliable.

**Question 21:** What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?  
**Answer:**  
It means we added too many variables — the model fits training data well but doesn’t generalize. Basically, it’s overfitting.

**Question 22:** Why is it important to scale variables in Multiple Linear Regression?  
**Answer:**  
Because different features may have different ranges, scaling keeps them on the same level and helps the model learn properly.

**Question 23:** What is Polynomial Regression?  
**Answer:**  
Polynomial Regression is used when the data forms a curve instead of a straight line. It adds powers of X like X², X³, etc., to make the line bend.

**Question 24:** How does Polynomial Regression differ from Linear Regression?  
**Answer:**  
Linear Regression makes a straight line, while Polynomial Regression can make curves because it adds squared or cubic terms.

**Question 25:** When is Polynomial Regression used?  
**Answer:**  
When the relationship between X and Y isn’t linear — like when the graph looks curved or U-shaped.

**Question 26:** What is the general equation for Polynomial Regression?  
**Answer:**  
Y = b₀ + b₁X + b₂X² + b₃X³ + ... + bₙXⁿ

**Question 27:** Can Polynomial Regression be applied to multiple variables?  
**Answer:**  
Yeah, we can apply it to multiple variables too. It’s called multivariate polynomial regression.

**Question 28:** What are the limitations of Polynomial Regression?  
**Answer:** 
- Can easily overfit.  
- Hard to interpret.  
- Bad at predicting beyond the data range.

**Question 29:** What methods can be used to evaluate model fit when selecting the degree of a polynomial?  
**Answer:**  
We can check:  
- Adjusted R²  
- Cross-validation score  
- AIC/BIC values  
- Visual plots of predictions

**Question 30:** Why is visualization important in Polynomial Regression?  
**Answer:**  
Because plotting helps you see if the curve really fits the data or if it’s too simple or too complex.

```python
# Question 31: How is polynomial regression implemented in Python?

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 6, 14, 28, 45])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```
